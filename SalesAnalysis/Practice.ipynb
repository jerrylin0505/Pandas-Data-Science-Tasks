### Read in updated dataframe

```python
all_data = pd.read_csv('all_data.csv')
all_data.head()
```

#### Drop rows of NAN

```python
nan_df = all_data[all_data.isna().any(axis=1)]
nan_df.head()

all_data = all_data.dropna(how='all')
```

#### Find "Or" and delete it

```python
all_data = all_data[all_data['Order Date'].str[0:2] != 'Or']
```

#### Convert columns to the correct type

```python
all_data["Quantity Ordered"] = pd.to_numeric(all_data["Quantity Ordered"])
all_data["Price Each"] = pd.to_numeric(all_data["Price Each"])
```

## Augment data with additional columns

### Task2: Add Month Column

```python
all_data["Month"] = all_data["Order Date"].str[0:2]
all_data["Month"] = all_data["Month"].astype('int32')
all_data.head()
```

### Task 3: Add a Sales column

```python
all_data['Sales'] = all_data["Quantity Ordered"] * all_data["Price Each"]

all_data.head()
```

### Task 4: Add a city column

```python
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

all_data["City"] = all_data["Purchase Address"].apply(lambda address: f"{get_city(address)} ({get_state(address)})")
all_data.head()
```

### Question 1:What was the best month for sales? How much was earned that month?

```python
result = all_data.groupby('Month').sum()
result
```

```python
import matplotlib.pyplot as plt

month = range(1,13)

plt.bar(month, result["Sales"])
plt.xticks(month)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month')
plt.show()
```

### Question 2: What city had the highest numbers of sales?

```python
result = all_data.groupby("City").sum()
result
```

```python
cities = [city for city, df in all_data.groupby("City")]

plt.bar(cities, result["Sales"])
plt.xticks(cities, rotation="vertical", size=8)
plt.ylabel('Sales in USD ($)')
plt.xlabel('City Name')
plt.show()
```

### Question 3: What time should we display advertisement to maximize likelihood of customer's buying product?

```python
all_data["Order Date"] = pd.to_datetime(all_data["Order Date"])
```

```python
all_data["Hour"] = all_data["Order Date"].dt.hour
all_data["Minute"] = all_data["Order Date"].dt.minute
```

```python
hours = [hour for hour, df in all_data.groupby("Hour")]

plt.plot(hours,all_data.groupby("Hour").count())
plt.xticks(hours)
plt.xlabel("Day Time", size=15)
plt.ylabel("Order Count", size=15)
plt.grid()
plt.show
```

### Question 4: What products are more often sold together

```python
#Order ID 相同則應該是同一筆訂單 ，所以要從相同的Order ID去找出哪些產品經常一起銷售

df = all_data[all_data["Order ID"].duplicated(keep=False)]
df["Grouped"] = df.groupby("Order ID")["Product"].transform(lambda x: ','.join(x))

df = df[["Order ID", "Grouped"]].drop_duplicates()

df.head(15)

```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df["Grouped"]:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))

for key, value in count.most_common(10):
    print(key, value)
```

### Question 5: What product sold the most? Why do you think it sold the most?

```python
product_group = all_data.groupby("Product")
product_group.sum()
quantity_ordered = product_group.sum()["Quantity Ordered"]
```

```python
products = [product for product, df in product_group]

plt.bar(products, quantity_ordered)
plt.ylabel("Quantity Odered")
plt.xlabel("Products")
plt.xticks(products, rotation="vertical", size=10)
plt.show()
```

```python
prices = all_data.groupby('Product').mean()["Price Each"]

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color="g")
ax2.plot(products, prices, color="b")


ax1.set_xlabel("Product Name")
ax1.set_ylabel("Quantity Ordered", color="g")
ax2.set_ylabel("Price($)", color="b")
ax1.set_xticklabels(products, rotation="vertical", size=10)

plt.show()
```
